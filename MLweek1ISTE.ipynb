```python
#question1
import numpy as np
import pandas as pd 
data = pd.read_csv('Titanic_ISTE.csv')
```

```python
from scipy.stats import chi2_contingency
crosstabresult_1=pd.crosstab(index=data['Survived'], columns=data['Pclass'])
chisq1=chi2_contingency(crosstabresult_1)
print('P-Value of Chi Square Test is' ,chisq1[1])

crosstabresult_2=pd.crosstab(index=data['Survived'], columns=data['Sex'])
chisq_2=chi2_contingency(crosstabresult_2)
print('P-Value of Chi Square Test is', chisq_2[1])

crosstabresult_3=pd.crosstab(index=data['Survived'], columns=data['Age'])
chisq_3=chi2_contingency(crosstabresult_3)
print('P-Value of Chi Square Test is', chisq_3[1])

crosstabresult_4=pd.crosstab(index=data['Survived'], columns=data['SibSp'])
chisq_4=chi2_contingency(crosstabresult_4)
print('P-Value of Chi Square Test is' ,chisq_4[1])

crosstabresult_5=pd.crosstab(index=data['Survived'], columns=data['Embarked'])
chisq_5=chi2_contingency(crosstabresult_5)
print('P-Value of Chi Square Test is' ,chisq_5[1])

crosstabresult_6=pd.crosstab(index=data['Survived'], columns=data['Fare'])
chisq_6=chi2_contingency(crosstabresult_6)
print('P-Value of Chi Square Test is' ,chisq_6[1])
```
